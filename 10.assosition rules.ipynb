```python
import pandas as pd
df=pd.read_excel("/content/Online retail.xlsx",sheet_name=0)
df.to_csv("content Online retail.csv",index=False)
```

```python
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns = ["items"]
transactions = df["items"].apply(lambda x: x.split(","))
transactions.head()
```

```python
#One-hot encode
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(te_array, columns=te.columns_)
basket.head()
```

```python
frequent_itemsets = apriori(basket, min_support=0.05, use_colnames=True)

print("Frequent itemsets found:", frequent_itemsets.shape)
frequent_itemsets.sort_values("support", ascending=False).head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print("Rules generated:", rules.shape)
rules.sort_values("lift", ascending=False).head(10)
```

```python
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Association Rules (Apriori)')
plt.show()
```

```python
rules.to_csv("association_rules_output.csv", index=False)
print("Rules saved to: association_rules_output.csv")
```

```python
#1.	What is lift and why is it important in Association rules?
#Lift measures how much more often two items (A and B) occur together than expected if they were independent.
#Lift helps us filter out â€œcoincidences.â€
# For example:If 90% of people buy milk, and 90% of people buy bread, they will co-occur often just by chance.
#Confidence alone may say â€œmilk â†’ breadâ€ is strong, but lift checks if the relationship is truly stronger than chance.
#2.	What is support and Confidence. How do you calculate them?
#Support tells us how frequently an itemset appears in the dataset.
#Confidence tells us how often item B is bought given that item A is bought.
#3.	What are some limitations or challenges of Association rules mining?
#Huge number of rules:Datasets can generate thousands of rules, many of which are trivial or not useful.
#Choosing thresholds:Picking meaningful min_support, min_confidence, and min_lift is tricky.
#Too high â†’ miss interesting rules. Too low â†’ too many irrelevant rules.
#Rare item problem:Rare but interesting items (e.g., luxury products) may be ignored because their support is too low.
#Interpretability:Not all strong rules make sense in real life (e.g., â€œpeople who buy caviar also buy waterâ€).
#Scalability:For very large datasets, Apriori can be computationally expensive since it scans the database multiple times.
```
